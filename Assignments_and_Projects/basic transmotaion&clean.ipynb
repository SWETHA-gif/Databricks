```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

data = [
  (1, "Alice",  "Engineering", 95000, "2021-03-15", "F"),
  (2, "Bob",    "Marketing",   72000, "2019-07-01", "M"),
  (3, "Carol",  "Engineering", 88000, "2020-11-20", "F"),
  (4, "David",  "HR",          61000, "2022-01-10", "M"),
  (5, "Eve",    "Marketing",   79000, "2018-05-25", "F"),
  (6, "Frank",  "Engineering", 102000,"2017-09-30", "M"),
  (7, "Grace",  "HR",          58000, "2023-02-14", "F"),
]
schema = StructType([
  StructField("id",         IntegerType(), False),
  StructField("name",       StringType(),  True),
  StructField("dept",       StringType(),  True),
  StructField("salary",     IntegerType(), True),
  StructField("join_date",  StringType(),  True),
  StructField("gender",     StringType(),  True),
])

df = spark.createDataFrame(data, schema)
```

```python
df.show()
```

```python
df.select("name", "dept", "salary").show()
```

```python
from pyspark.sql.functions import col

df.select(col("name").alias("NAME"), col("salary")).show()
```

```python
from pyspark.sql.functions import col, upper

df.select(
    col("name"),
    (col("salary") / 12).alias("monthly_salary"),
    upper(col("dept")).alias("dept_upper")
).show()
```

```python
#withcolumn
df2 = df.withColumn("monthly_salary", col("salary") / 12)
df2.select("name", "salary", "monthly_salary").show(3)
```

```python
df2.show()
```

```python
from pyspark.sql.functions import col, to_date, upper

df2 = df.withColumn("join_date", to_date(col("join_date"), "yyyy-MM-dd")) \
        .withColumn("name_upper", upper(col("name"))) \
        .withColumn("salary", col("salary").cast("double"))

display(df2)
```

```python
df.drop("gender").show(3)
```

```python
df.withColumnRenamed("dept", "department") \
  .withColumnRenamed("salary", "annual_salary") \
  .show(3)
```

```python
df.select(
    col("name"),
    col("dept").alias("department"),
    col("salary").alias("annual_salary")
).show(3)
```

```python
df.filter(col("salary") > 80000).show()
df.where(col("salary") > 80000).show()
```

```python
# AND
df.filter((col("dept") == "Engineering") & (col("salary") > 90000)).show()

# OR
df.filter((col("dept") == "HR") | (col("salary") < 65000)).show()
```

```python
df.orderBy("salary").show()
df.sort("salary").show()    
```

```python
from pyspark.sql.functions import count,avg,max
df.groupBy("dept").agg(
    count("id").alias("headcount"),
    avg("salary").alias("avg_salary"),
    max("salary").alias("max_salary")
).show()
```
