```python
df.show()
```

```python
df.select("Name", "Designation", "Salary").show()
```

```python
# Filter data Employees with salary > 500000 Employees from Bangalore

from pyspark.sql.functions import col
df.filter(col("Salary")>500000).show()
```

```python
# 4. Count records Total number of employees
total_employee = df.count()
total_employee
```

```python
# 5. Distinct values Get all unique cities
from pyspark.sql.functions import col
df.select(col("Location")).distinct().show()
```

```python
# Sort employees by salary (high → low)
from pyspark.sql.functions import col
df.orderBy(col("Salary").desc()).show()
```

```python
# Rename name → employee_name
from pyspark.sql.functions import col
df.select(col("Name").alias("employee_name")).show()
```

```python
# Add column salary_in_lakhs (salary / 100000)
from pyspark.sql.functions import col
df.withColumn("Salary_in_lakhs", col("Salary")/100000).show()
```

```python
from pyspark.sql.functions import col, avg
df.groupBy("Location").agg(avg(col("Salary")).alias("avg_salary")).show()
```
