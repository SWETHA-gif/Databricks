```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

orders_data = [
    (101, 1, "2023-01-10", 2500, "Completed"),
    (102, 2, "2023-02-15", 1800, "Pending"),
    (103, 3, "2023-03-01", 3200, "Completed"),
    (104, 1, "2023-03-20", 1500, "Cancelled"),
    (105, 4, "2023-04-05", 2700, "Completed"),
]

orders_schema = StructType(
    [ StructField("order_id", IntegerType(), True),
      StructField("customer_id", IntegerType(), True),
      StructField("order_date", StringType(), True),
      StructField("amount", IntegerType(), True),
      StructField("status", StringType(), True)
    ]
)
```

```python
customers_data = [
    (1, "Swetha", "Bangalore", 25),
    (2, "Ravi", "Mysore", 30),
    (3, "Anu", "Chennai", 28),
    (4, "Kiran", "Hyderabad", 35),
    (5, "John", "Delhi", 40),
]

customers_cols = ["customer_id", "name", "city", "age"]
```

```python
products_data = [
    (201, "Laptop", "Electronics", 50000),
    (202, "Phone", "Electronics", 20000),
    (203, "Shoes", "Fashion", 3000),
]

products_cols = ["product_id", "product_name", "category", "price"]
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

order_items_data = [
    (101, 201, 1),
    (101, 203, 2),
    (102, 202, 1),
    (103, 201, 1),
    (104, 203, 3),
]
order_items_schema = StructType(
    [ StructField("order_id", IntegerType(), True),
      StructField("product_id", IntegerType(), True),
      StructField("quantity", IntegerType(), True)
    ]
)
```

```python
# Creating dataframe for orders data
df_orders = spark.createDataFrame(orders_data, orders_schema)
df_orders.write.mode("overwrite").csv("/Volumes/dataframeassigment/e_commerce/ecommerce/Orders", header=True)
                                      
# creating dataframe for customers data
df_customers = spark.createDataFrame(customers_data, customers_cols)
df_customers.write.mode("overwrite").json("/Volumes/dataframeassigment/e_commerce/ecommerce/Customers")

# creating dataframe for products data
df_products = spark.createDataFrame(products_data,products_cols)
df_products.write.mode("overwrite").parquet("/Volumes/dataframeassigment/e_commerce/ecommerce/Products")

# creating dataframe for order_items data
df_order_items = spark.createDataFrame(order_items_data, order_items_schema) 
df_order_items.write.mode("overwrite").csv("/Volumes/dataframeassigment/e_commerce/ecommerce/Order_Items", header=True)
```

```python
# Read the orders data
df_orders = spark.read.csv("/Volumes/dataframeassigment/e_commerce/ecommerce/Orders", header=True, inferSchema=True)


# Read the customers data
df_customers = spark.read.json("/Volumes/dataframeassigment/e_commerce/ecommerce/Customers")


# Read the products data
df_products = spark.read.parquet("/Volumes/dataframeassigment/e_commerce/ecommerce/Products")


# Read the order_items data
df_order_items = spark.read.csv("/Volumes/dataframeassigment/e_commerce/ecommerce/Order_Items", header=True, inferSchema=True)

```

```python
print("orders schema: ")
df_orders.printSchema()
print("customers schema: ")
df_customers.printSchema()
print("products schema: ")
df_products.printSchema()
print("order_items schema: ")
df_order_items.printSchema()
```

```python
# LEVEL 1 -BASIC -- 2. Convert order_date column to DateType.
from pyspark.sql.functions import col
df_dateConversion = df_orders.withColumn("order_date", col("order_date").cast("date"))
df_dateConversion.display()
```

```python
# LEVEL 1 -BASIC -- 3. Filter only orders where status = 'Completed'.
from pyspark.sql.functions import col
df_orders.filter(col("status")=="Completed").display()
```

```python
# LEVEL 1 -BASIC -- 4. Select customer name and city from customers DataFrame.
from pyspark.sql.functions import col
df_customers.select("name", "city").show()
```

```python
# LEVEL 2 - INTERMEDIATE -- 5. Find total number of orders per customer.

from pyspark.sql.functions import count

df_orders.groupBy("customer_id").agg(count("*").alias("total_orders")).join(df_customers, "customer_id").select("customer_id", "name", "total_orders").display()
```

```python
# LEVEL 2 - INTERMEDIATE --6. Find total revenue (sum of order amount).

from pyspark.sql.functions import col, sum
df_products.agg(sum(col("price")).alias("total_revenue")).display()
```

```python
# LEVEL 2 - INTERMEDIATE -- 7. Find average order value.

from pyspark.sql.functions import col, avg
df_orders.agg(avg(col("amount")).alias("avg_order_value")).display()
```
