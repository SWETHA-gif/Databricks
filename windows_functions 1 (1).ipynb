```python
%sql
CREATE CATALOG IF NOT EXISTS practice1;
CREATE SCHEMA IF NOT EXISTS practice1.test;

CREATE TABLE IF NOT EXISTS practice1.test.emp_window (
    emp_id INT,
    name STRING,
    department STRING,
    salary INT
);


INSERT INTO practice1.test.emp_window VALUES
(1, 'A', 'IT', 5000),
(2, 'B', 'IT', 6000),
(3, 'C', 'IT', 6000),
(4, 'D', 'HR', 4000),
(5, 'E', 'HR', 4000),
(6, 'F', 'HR', 3000);
```

```python
%sql
SELECT department, SUM(salary)
FROM  practice1.test.emp_window
GROUP BY department;
```

```python
%sql
SELECT 
    name,
    salary,
    SUM(salary) OVER () AS total_salary
FROM practice1.test.emp_window;
```

```python
%sql
SELECT *,
       ROW_NUMBER() OVER (PARTITION BY department ORDER BY salary DESC) AS rn
FROM practice1.test.emp_window;
```

```python
%sql
SELECT *,
       RANK() OVER (PARTITION BY department ORDER BY salary DESC) AS rnk
FROM practice1.test.emp_window;
```

```python
%sql
SELECT *,
       DENSE_RANK() OVER ( ORDER BY salary DESC) AS drnk
FROM practice1.test.emp_window;
```

```python
%sql
SELECT *
FROM (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY department ORDER BY salary DESC) AS rn
    FROM practice1.test.emp_window 
) t
WHERE rn <= 3;
```

```python
%sql
SELECT *
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY department ORDER BY salary DESC) AS rnk
    FROM  practice1.test.emp_window 
) t
WHERE rnk = 3;
```

```python
%sql
SELECT name, salary,
       ROW_NUMBER() OVER (ORDER BY salary DESC) AS row_num,
       RANK() OVER (ORDER BY salary DESC) AS rank,
       DENSE_RANK() OVER (ORDER BY salary DESC) AS dense_rank
FROM practice1.test.emp_window;
```

```python
%sql
SELECT 
    name,
    salary,
    LAG(salary) OVER (ORDER BY salary) AS prev_salary
FROM practice1.test.emp_window;
```

```python
%sql
SELECT 
    name,
    salary,
    salary - LEAD(salary) OVER (ORDER BY salary) AS diff
FROM practice1.test.emp_window;
```

```python
%sql
SELECT *
    ,salary - LAG(salary) OVER (ORDER BY salary) AS diff
FROM practice1.test.emp_window;
```

```python
%sql
SELECT 
    name,
    salary,
    SUM(salary) OVER (ORDER BY name) AS running_total
FROM practice1.test.emp_window;
```
