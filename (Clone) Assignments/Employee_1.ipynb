```python
df = spark.table("dataframeassigment.default.employees_1")
df.show()
```

```python
df.display()
```

```python
"""
1.  Display only the first 4 rows of the employees DataFrame.
2.  Show all rows without truncating any column values.
3.  Select only name, dept, and salary from the DataFrame.
4.  Select name and a new column annual_bonus which is 10% of salary.
5.  Add a column tax which is 20% of the salary.
6.  Add a column experience based on join_date year: before 2019 = 'Senior', 2019–2021 = 'Mid', after 2021 = 'Junior'.
7.  Drop the gender column from the DataFrame.
8.  Rename dept to department.
9.  Rename salary to annual_salary and id to emp_id in one chain.
10. Rename all columns to uppercase at once.
11. Filter all employees in the HR department.
12. Find all active employees (is_active = True) with a rating of 5.
13. Find employees NOT in Engineering or Marketing.
14. Sort the DataFrame by salary in descending order.
15. Sort by dept ascending, then by salary descending within each dept.
16. Count the number of employees in each department.
17. For each department, find the total salary, average salary, and max salary
"""
```

```python
# 1.Display only the first 4 rows of the employees DataFrame.
df.limit(4).show()
```

```python
# 2.Show all rows without truncating any column values.
df.show(truncate=False)
```

```python
# Select only name, dept, and salary from the DataFrame.
df.select("First Name", "Team", "Salary").show()
```

```python
# 4.  Select name and a new column annual_bonus which is 10% of salary.
df.select("First Name", "Salary", (df["Salary"]/10).alias("annual_bonus")).show()
```

```python
from pyspark.sql.functions import col
df.schema["Salary"].dataType
```

```python
# 5. Add a column tax which is 20% of the salary
from pyspark.sql.functions import col
df.withColumn("Tax", col("Salary").cast("int")*0.20).display()
```

```python

# 6. Add a column experience based on join_date year: before 2019 = 'Senior', 2019–2021 = 'Mid', after 2021 = 'Junior
from pyspark.sql.functions import col, when, year
df.withColumn(
    "experience",
    when(year(col("Start Date")) < 2019, "Senior")
    .when((year(col("Start Date")) >= 2019) & (year(col("Start Date")) <= 2021), "Mid")
    .otherwise("Junior")
).display()
```

```python
# 7. Drop the gender column from the DataFrame.
df.drop("Gender").display()
```

```python
# 8. Rename dept to department.
df.withColumnRenamed("Team", "Department").show()
```

```python
# 9. Rename salary to annual_salary and id to emp_id in one chain.
df.withColumnRenamed("Salary", "annual_salary")\
.withColumnRenamed("First Name", "Employee Name").display()
```

```python
# 10. Rename all columns to uppercase at once.

df.toDF(*[col_name.upper() for col_name in df.columns]).display()
```

```python
# 11. Filter all employees in the HR department.
from pyspark.sql.functions import col
df.filter(col("Team") == "Human Resources").display()
```

```python
# 12. Find all active employees (is_active = True) with a rating of 5.
from pyspark.sql.functions import col
df.filter(
    (col("Senior Management") == True) & (col("Bonus %") >=10)
).display()
```

```python
# 12. Find all active employees (is_active = True) with a rating of 5.
# used the columns in the given df

from pyspark.sql.functions import col
df.filter((col("Senior Management") == True) & (col("Bonus %") >=18)).display()
```

```python
# 13. Find employees NOT in Engineering or Marketing.
from pyspark.sql.functions import col
df.filter((col("Team")!="Engineering") & (col("Team")!="Marketing")).display()
```

```python
# Sort the DataFrame by salary in descending order.
from pyspark.sql.functions import col
df.orderBy(col("Salary").desc()).display()
```

```python
# Sort by dept ascending, then by salary descending within each dept.
from pyspark.sql.functions import col
df.orderBy(col("Team"), col("Salary").desc()).display()
```

```python
# 16. Count the number of employees in each department.

df.groupBy("Team").count().display()

```

```python
# 17. For each department, find the total salary, average salary, and max salary
from pyspark.sql.functions import sum, avg, max

df.groupBy("Team").agg(
    sum("Salary").alias("total_salary"),
    avg("Salary").alias("avg_salary"),
    max("Salary").alias("max_salary")
).display()
```
